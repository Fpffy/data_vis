```python
austin = pd.read_csv('austin.txt', sep=',')
display(austin.head(n=2))
```

```python
# 1
fig_1, ax_1 = plt.subplots()
ax_1.plot(austin['month_avg'], color='magenta', marker='x', linestyle='--')
ax_1.set_title('Monthly average temperature in Austin')
ax_1.set_xlabel('Month')
ax_1.set_ylabel('Average temperature [Fahrenheit degrees]')
ax_1.set_xticks(range(0, 12))
ax_1.set_xticklabels(austin['month'], rotation=45)
plt.show()
```

```python
# 2
fig_1, ax_2 = plt.subplots(3, 2, sharey=True)
ax_2[1, 1].plot(austin['month_std'])
plt.show()
```
