## Quantitative comparisons and statistical visualizations

#### 2. Histograms
    2.1 Histograms of two variable

    4.2 Diffrent ways to change color

```python
climate = pd.read_csv('climate.csv', index_col='year')
medals = pd.read_csv('medals.csv', index_col='country')
male = pd.read_csv('male.csv')
female = pd.read_csv('female.csv')
austin = pd.read_csv('austin.csv', index_col='month')
seattle = pd.read_csv('seattle.csv', index_col='month')
```

```python
# 1
fig_1, ax_1 = plt.subplots()
ax_1.bar(medals.index, medals['gold'], label='Gold', color='yellow')
ax_1.bar(medals.index, medals['silver'], label='Silver',
         bottom=medals['gold'], color='gray')
ax_1.bar(medals.index, medals['bronze'], label='Bronze',
         bottom=medals['gold']+medals['silver'], color='brown')
ax_1.set_xticklabels(labels=medals.index, rotation=90)
ax_1.set_ylabel('Number of medals')
ax_1.legend()
plt.show()
```

```python
# 2
fig_2, ax_2 = plt.subplots()
ax_2.hist([male['height'], female['height']], bins=30, label=['Men' ,'Women'],
          histtype='step')
ax_2.set_xlabel('Height')
ax_2.set_ylabel('# of observations')
ax_2.legend()
plt.show()
```

```python
# 3.1
fig_3, ax_3 = plt.subplots()
ax_3.bar('Men', male['weight'].mean(), yerr=male['weight'].std(), label='Men')
ax_3.bar('Woman', female['weight'].mean(), yerr=female['weight'].std(), label='Women')
ax_3.set_xlabel('Gender')
ax_3.set_ylabel('Weight')
plt.show()
```

```python
# 3.2
fig_4, ax_4 = plt.subplots()
ax_4.errorbar(austin.index, austin['month_avg'], yerr=austin['month_std'], label='Austin')
ax_4.errorbar(seattle.index, seattle['month_avg'], yerr=seattle['month_std'], label='Seattle')
ax_4.set_xlabel('Month')
ax_4.set_ylabel('Temperature [Fahrenheit]')
ax_4.legend(loc='upper left')
plt.show()
```

```python
# 3.3
fig_5, ax_5 = plt.subplots()
ax_5.boxplot([male['height'], female['height']])
ax_5.set_xticklabels(['Men', 'Women'])
ax_5.set_ylabel('Height')
plt.show()
```

```python
#4
fig_6, ax_6 = plt.subplots()
ax_6.scatter(climate['co2'], climate['temp'], c=climate.index)
ax_6.set_xlabel('CO2 level')
ax_6.set_ylabel('Relative temperature')

plt.show()
```
