## Preparing your figures to share with others

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

1. Height mean and std in function of single sport
2. Two plots with diffrent style
3. Figure size
4. Save png, diffrent dpi
5. Save jpeg, diffrent compression

```python
olympic = pd.read_csv('olympic.csv')
olympic = olympic.dropna()
```

```python
sport_col = olympic['Sport']
sports = sport_col.unique()
```

```python
plt.style.use('ggplot')

fig_1, ax_1 = plt.subplots()

for sport in sports:
    heights = olympic[olympic['Sport'] == sport]['Height']
    ax_1.bar(sport, heights.mean(), yerr=heights.std())
    
fig_1.set_size_inches([20, 10])
ax_1.set_xticklabels(sports, rotation=90)
ax_1.set_ylabel('Height [cm]')
fig_1.savefig('heights_gg_300.png', dpi=300)
fig_1.savefig('heights_gg_100.png', dpi=100)
plt.show()
```

```python
plt.style.use('default')

fig_1, ax_1 = plt.subplots()

for sport in sports:
    heights = olympic[olympic['Sport'] == sport]['Height']
    ax_1.bar(sport, heights.mean(), yerr=heights.std())
    
fig_1.set_size_inches([20, 10])
ax_1.set_xticklabels(sports, rotation=90)
ax_1.set_ylabel('Height [cm]')
plt.savefig('height_50.jpeg', compression=50, dpi=300)
plt.show()
```
